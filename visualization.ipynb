```python
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_csv('194903-202105_Bottle.csv', encoding='latin1', low_memory=False)
```

```python
df['Year'] = df['Depth_ID'].str.split('-').str[1].str[:2].astype(float)
df['Year'] = df['Year'].apply(lambda x: 1900 + x if x >= 49 else 2000 + x)
```

```python
features = [
    'T_degC', 'Salnty', 'O2ml_L', 'NO3uM',
    'PO4uM', 'ChlorA', 'Depthm', 'Year'
]
```

```python
df_clean = df[features].copy()
df_clean = df_clean.dropna(subset=['O2ml_L', 'NO3uM', 'T_degC', 'Salnty', 'Year'])
```

```python
df[["T_degC", "Salnty", "O2ml_L", "Depthm"]].isnull().sum()
```

```python
px.scatter(
    df_clean.sample(5000),
    x='T_degC',
    y='O2ml_L',
    color='Depthm',
    title='Temperature vs Oxygen (Exploration)'
)
```

```python
px.scatter(
    df_clean.sample(5000),
    x='Depthm',
    y='O2ml_L',
    title='Depth vs Oxygen'
)
```

```python
px.scatter(
    df_clean.sample(5000),
    x='NO3uM',
    y='O2ml_L',
    title='Nitrate vs Oxygen'
)
```

```python
df_year = df_clean.groupby('Year')['O2ml_L'].mean().reset_index()

px.line(
    df_year,
    x='Year',
    y='O2ml_L',
    title='Average Oxygen Over Time'
)
```
